## Soru 1:
* Kaggle'in 2020 yılı için üyeleriyle yaptığı anket sonuçlarını içeren ekteki "kaggle_survey_2020_responses.csv" dosyasını 2. satırı atlayarak ```survey``` adlı DataFrame'e okuyun.
* İlk sütunu silin.
* Q1, Q2, Q3, Q4, Q5 ve Q6 sütun isimlerini sırasıyla Age, Gender, Country, Education, Occupation ve Coding Experience olacak şekilde değiştirin.
* Education, Occupation ve Coding Experience sütunlarından herhangi birisisinde ```NaN``` değeri olan satırları siliniz. Kaç satır silindi?
* En son oluşan DataFrame'in satır ve sütun sayısını yazdırın.

````python
#Gerekli kütüphanelerin dahil edilmesi
import pandas as pd
import numpy as np
import random as rd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
````

````python
#Csv dosyasının 2. satır atlanarak okunması
survey = pd.read_csv("kaggle_survey_2020_responses.csv", index_col = False, skiprows=[1])
#İlk sütunun silinmesi
survey.drop(["Time from Start to Finish (seconds)"], axis = 1, inplace = True)

#İlgili sütunların isimlerinin değiştirilmesi
survey.rename({"Q1":"Age", "Q2":"Gender", "Q3":"Country", 
                                  "Q4":"Education", "Q5":"Occupation", "Q6":"Coding Experience"}, axis = 1, inplace = True)

rowCount = survey.shape[0]
survey.dropna(subset = ["Education", "Occupation", "Coding Experience"], how = "all", inplace = True)

print("Silinen satır sayısı :", rowCount - survey.shape[0])
print("Survey DataFrame'inin en son satır ve sütun sayıları")
print("Satır :", survey.shape[0], "\nSütun :", survey.shape[1])
````

````python
group_age = survey.groupby(['Age'], observed=True)
age_interval = group_age.groups.keys()
age_interval = group_age.size()

plot = age_interval.plot.pie(y=0,autopct='%1.1f%%',title="Ankete katılanların yaş aralıklarına göre dağılımı",
                    figsize=(10, 10),legend=True)
````

## Soru 8:
* Ortalama yıllık maaşı en yüksek olan meslek (```Occupation```) nedir? ```Data Scientist``` kaçıncı sıradadır? **Not:** Aralık olarak verilen maaşların orta noktasını ***(aralık olmayan değerler için ifadedeki değeri)*** maaş değeri olarak alıp ```ConvertedComp``` isimli yeni bir sütun ekleyiniz.
* ```Data Scientist``` olarak çalışanların ülke bazında ortalama yıllık maaşlarını büyükten küçüğe sıralayın, ilk 10 ülkeyi ve bu ülkelerdeki ortalama yıllık ```Data Scientist``` maaşlarını listeleyin.
* 250’den fazla kişi çalışan kurumdakilerin ortalama maaşları ne kadardır?
* Hesaplanan maaş (```ConvertedComp```) sütununu kullanarak ülke bazlı ortalama, standart sapma ve medyan değerlerinden oluşan bir dataframe oluşturun ve listeleyin. Türkiye kaçıncı sıradadır ve değeri kaçtır?

````python
survey['Q24'] = survey['Q24'].str.replace(',','')
survey['Q24'] = survey['Q24'].str.replace('$','')
survey['Q24'].replace({'> 500000':'1000000'}, inplace=True) #Aşağıda sağ taraftaki değer ile sol taraftaki değerin 
                                                            #toplamının yarısı alındığı için tekrardan kendisine eşit olması
                                                            #sebebiyle 1000000 ile değiştirilmiştir.
survey['Q24_Left']=survey.Q24.str.split('-').str[0]
survey['Q24_Right']=survey.Q24.str.split('-').str[1]
survey['Q24_Left'].fillna(0, inplace=True)
survey['Q24_Right'].fillna(0, inplace=True)

#Maaş ortalamasının bulunması
survey['ConvertedComp'] = (survey.Q24_Right.values.astype('int32') + survey.Q24_Left.values.astype('int32'))/2
````

````python
mostSalaryOccu = survey.groupby('Occupation')['ConvertedComp'].agg(['mean'])
mostSalaryOccu = mostSalaryOccu.sort_values(by='mean',ascending=False)
print('Ortalama yıllık maaşı en yüksek meslek :', mostSalaryOccu.idxmax())
````

````python
mostSalaryOccu["Rank"] = list(range(len(mostSalaryOccu.index)))
mostSalaryOccu["Rank"] = mostSalaryOccu["Rank"] + 1
mostSalaryOccu2 = mostSalaryOccu.copy()
mostSalaryOccu2.reset_index(inplace=True)
rank = mostSalaryOccu2[mostSalaryOccu2.Occupation == 'Data Scientist']
print("Data Scientist'lerin sırası : ", rank['Rank'].values, "\n")
````

````python
dataScientist = survey[survey.Occupation == 'Data Scientist']
groupedCountry = dataScientist.groupby('Country')['ConvertedComp'].mean().sort_values(ascending=False).to_frame()
print('Yıllık ortalama maaşların ülke bazında ortalamasının listelenmesi: \n', groupedCountry, '\n')
print('Yıllık ortalama maaşı en fazla olan ilk 10 ülke :\n', groupedCountry[0:10], '\n')
````

````python
survey250 = survey[(survey.Q20 != '0-49 employees') & (survey.Q20 != '50-249 employees')]
print("250 kişiden fazla çalışanı olan kurumlarda çalışanların ortalama maaşı :", survey250['ConvertedComp'].mean(), '\n')
````

````python
compCountry = survey.groupby('Country')['ConvertedComp'].agg(['mean','std','median'])
print('Ülke bazlı ortalama, standart sapma ve medyandan oluşan dataframe :')
compCountry
````

````python
compCountry = compCountry.reset_index()
compCountry["Rank"] = list(range(len(compCountry.index)))
compCountry["Rank"] = compCountry["Rank"] + 1
compCountry = compCountry.copy()
compCountry.reset_index(inplace=True)
rank = compCountry[compCountry.Country == 'Turkey']
print("Türkiye'nin bulunduğu sıra : ",rank['Rank'].values, "\n")
print("Türkiye'nin değerleri : \n", rank)
````

## Soru 10:

## Soru 13:
* https://download.geonames.org/export/dump/countryInfo.txt adresindeki veriyi:
    * ilk 49 satırını ihmal ederek
    * yalnızca ```#ISO, Country, neighbours``` sütunlarını kullanarak 
    * ```Country``` sütununu indeks sütun olarak  
    
okuyarak __country_info__ adlı DataFrame'e aktarınız.
* ```#ISO``` sütununun ismini ```Country Code``` olarak değiştiriniz.
* ```survey``` ve ```country_info``` adlı DataFrame'leri ülke bazında birleştirip ```merged_survey``` adında yeni bir dataframe'e aktarınız.

````python
#Sisteme upload edilen txtnin okunup "\t" aracılığıyla verilerin ayrıştırılması ve sadece ilgili sütunların df'me eklenmesi
country_info = pd.read_csv("countryInfo.txt",sep="\t",skiprows=49,usecols=["#ISO","Country","neighbours"], na_filter=False)

#"Country"nin index olarak atanması
country_info.set_index("Country", inplace = True)

#İstenen kolon isminin güncellenmesi
country_info.rename(columns={"#ISO":"Country Code"}, inplace = True)

#İstenen dataframelerin merge edilmesi
merged_survey = survey.merge(country_info, left_on = "Country", right_on = "Country")
````

## Soru 14:
Almanya ve komşu ülkeleri için ankete katılanlardan doktoralıların ülke bazında yüzdelerini, ülke isimlerine göre alfabetik sırada listeleyiniz.

````python
#Almanya ve komşu ülkelerinin seçilmesi
merged_survey=merged_survey[(merged_survey.Country=="Germany")|(merged_survey.Country=="Denmark")|(merged_survey.Country=="Poland")
                            |(merged_survey.Country=="Czech Republic")|(merged_survey.Country=="Austria")|
                            (merged_survey.Country=="Switzerland")|(merged_survey.Country=="France")|
                            (merged_survey.Country=="Luxemburg")|(merged_survey.Country=="Belgium")|
                            (merged_survey.Country=="Netherlands")]

#Doktora seviyesinde eğitime sahip olanların seçilmesi
merged_survey=merged_survey[merged_survey.Education == "Doctoral degree"]

#Ülke bazında doktora seviyesindeki kişilerin sayılarının toplanması
percent_doctoral=merged_survey.groupby("Country")[["Education"]].count()

#Toplanan sayılar ile birlikte ülke bazında gruplanmasıyla elde edilen sonuçtan yüzde hesaplanması
percent_doctoral['Percent'] = (percent_doctoral['Education'] / 
                  percent_doctoral['Education'].sum()) * 100

#Ekrana yazdırılması için gereksiz sütun olan Education'un silinmesi ve sonrasında alfabetik olarak sıralanması
percent_doctoral.drop("Education", axis='columns', inplace=True)
percent_doctoral.sort_values(by="Country", ascending=True)
````
